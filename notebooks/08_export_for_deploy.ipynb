```python
import pandas as pd
from sqlalchemy import create_engine
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import config

engine = create_engine(
    f"mysql+pymysql://{config.DB_CONFIG['user']}:{config.DB_CONFIG['password']}"
    f"@{config.DB_CONFIG['host']}:{config.DB_CONFIG['port']}/{config.DB_CONFIG['database']}"
)

pd.read_sql("SELECT * FROM stocks",            engine).to_csv("../data/processed/stocks.csv",            index=False)
pd.read_sql("SELECT * FROM stock_volatility",  engine).to_csv("../data/processed/stock_volatility.csv",  index=False)
pd.read_sql("SELECT * FROM stock_risk_return", engine).to_csv("../data/processed/stock_risk_return.csv", index=False)

# Also export seasonality + trend data
pd.read_sql("SELECT ticker, trade_date, close_price, ma_50, ma_200 FROM stock_features", engine).to_csv("../data/processed/stock_features_lite.csv", index=False)
pd.read_sql("SELECT market, MONTH(trade_date) AS mn, ROUND(AVG(daily_return)*100,4) AS avg_ret FROM stock_features WHERE daily_return IS NOT NULL GROUP BY market, MONTH(trade_date)", engine).to_csv("../data/processed/seasonality.csv", index=False)

print("Done")
```
